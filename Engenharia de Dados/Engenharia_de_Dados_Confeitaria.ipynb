```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("/content/relatorio_items_vendas20251016-1-rwrl26.xlsx")
df.head()
```

```python
df.info()
```

Colunas estão em formatos consistentes, com as colunas categórias (Produto e Categoria) com o tipo object e as colunas numéricas (Quantidade vendida até margem) com os tipos int e float. Além disso, todas as colunas não possuem valores nulos.

Entretanto, acredito que seria melhor deixar as colunas "Custo Médio e Custo Total" como float, pois elas são valores monetários.

```python
df[['Custo Médio', 'Custo Total']] = df[['Custo Médio', 'Custo Total']].astype(float)
df.info()
```

# Normalização - Produtos

```python
df['Produto'].unique()
```

Alguns produtos possuem espaços em excesso.

```python
# Removendo espaços e deixando tudo em maiusculo
df['Produto_corrigido'] = df['Produto'].str.strip().str.upper()
```

```python
df['Produto_corrigido'].unique()
```

Produto "Naked Red Velvet" possui erros ortográficos que podem ser corrigidos.

```python
df.loc[df['Produto_corrigido'] == 'NAKED - RED VELVET (10 FATIAS - ZERO AÇÚCAR)']
```

```python
df.loc[df['Produto_corrigido'] == 'NAKED. RED VELVET - 10 FATIAS - ZERO']
```

```python
df['Produto_corrigido'] = df['Produto_corrigido'].str.replace('NAKED. RED VELVET - 10 FATIAS - ZERO', 'NAKED - RED VELVET (10 FATIAS - ZERO AÇÚCAR)')
df.loc[df['Produto_corrigido'] == 'NAKED - RED VELVET (10 FATIAS - ZERO AÇÚCAR)']
```

```python
pd.set_option('display.max_rows', None)
df['Produto_corrigido'].value_counts()
```

```python
# Corrigindo alguns erros de espaço em alguns produtos

df['Produto_corrigido'] = df['Produto_corrigido'].str.replace('FATIA -CHOCOMELO (150G)', 'FATIA - CHOCOMELO (150G)')
df['Produto_corrigido'] = df['Produto_corrigido'].str.replace('FOCACCIA- PESTO & FRANGO	', 'FOCACCIA - PESTO & FRANGO	')
df['Produto_corrigido'] = df['Produto_corrigido'].str.replace('POTE DA FELICIDADE -MOUSE DE CHOCOLATE BRANCO E MORANGO', 'POTE DA FELICIDADE - MOUSE DE CHOCOLATE BRANCO E MORANGO')
df['Produto_corrigido'] = df['Produto_corrigido'].str.replace('BOLO COM RASPAS DE CHOCOLATE(8 FATIAS) ZERO AÇÚCAR', 'BOLO COM RASPAS DE CHOCOLATE (8 FATIAS) ZERO AÇÚCAR')
```

# Normalização - Categoria

```python
df['Categoria'].value_counts()
```

Assim como a coluna 'Produto', a coluna 'Categoria' apresenta alguns erros de grafia e nomes repetidos que podem ser corrigidos.

```python
# Removendo espaços no começa o no final de cada nome e deixando tudo em maiuscúlo
df['Categoria_corrigido'] = df['Categoria'].str.strip().str.upper()
```

```python
df['Categoria_corrigido'] = df['Categoria_corrigido'].replace('CESTA E PRESENTES', 'CESTAS E PRESENTES')
```

```python
df['Categoria_corrigido'].unique()
```

```python
# Corrigindo alguns erros de espaço e sinalização
df['Categoria_corrigido'] = df['Categoria_corrigido'].str.replace('LANCHES - SEM GLÚTEN E SEM LACTOSE !', 'LANCHES - SEM GLÚTEN E SEM LACTOSE')
df['Categoria_corrigido'] = df['Categoria_corrigido'].str.replace('TORTAS - AGENDAMENTO 24H  - RETIRADA EM LOJA -', 'TORTAS - AGENDAMENTO 24H  - RETIRADA EM LOJA')
```

```python
df['Categoria_corrigido'].unique()
```

```python
# Inserido os dados normalizados nas colunas originais e removendo as colunas de teste.
df['Produto'] = df['Produto_corrigido']
df['Categoria'] = df['Categoria_corrigido']
df = df.drop(columns=["Produto_corrigido", "Categoria_corrigido"])
```

## Atenção

Alguns produtos possuem exatamente o mesmo nome, mas estão em categorias diferentes, recomenda-se encaixar esses produtos em um categoria única para melhor análise.

```python
duplicados = df.groupby("Produto")["Categoria"].nunique()
produtos_multicat = duplicados[duplicados > 1].index

df_multicat = df[df["Produto"].isin(produtos_multicat)].sort_values(["Produto", "Categoria"])

tabela_resumo = df_multicat[["Produto", "Categoria", "Preço Médio"]].sort_values(["Produto", "Categoria"])

tabela_resumo
```

## Verificação Relacional

```python
# Verificando se as colunas quantidade e preço médio resultam nos mesmos valores do faturamento
df['Faturamento_calculado'] = df['Quantidade Vendida'] * df['Preço Médio']
df[['Faturamento_calculado', 'Faturamento Total']]
```

Relação entre quantidade vendida, preço médio e faturamento total aparentam ter uma relação consistente.

```python
# Verificar se o lucro também é consistente
df['Lucro_calculado'] = df['Faturamento Total'] - (df['Custo Médio'] * df['Quantidade Vendida'])
df[['Lucro_calculado', 'Lucro']]
```

A relação entre lucro, quantidade vendida, preço e custo médio também apresentem consistências.
